# Exemples Àlgebra

```python
import numpy as np
```

Exemples de vectors (array)

```python
v=np.array([1,0,0])
u=np.array([0,-2,0])
```

```python
print(u)
```

Operacions amb vectors (suma, resta, producte escalar i vectorial)

```python
w1=u+v
print(w1)
w2=u-v
print(w2)
```

```python
a=np.dot(u,v) #np.dot es per fer producte escalar
print(a)
```

```python
b=np.cross(u,v) #np.cross es per producte vectorial
print(b)
```

Exemple amb Matrius

```python
A=np.array([[2,4],[1,3]])
print(A)
```

```python
e=A[0,1] # 0 es la fila i 1 es la columna
print(e)
```

```python
v=A[0,:] #els : es per agafar tots vectors de la fila 0. És com un comodin
print(v)
```

```python
u=A[:,1] # agafa tots els valors de la columna 1
print(u)
```

Diagonalitzem matrius (busquem valors i vectors propis)

```python
A=np.array([[2,2],[1,3]])
```

```python
Valors,Vectors = np.linalg.eig(A) # np.linalg.eig8) per veure valors i vectos propis
print("Valors propis", Valors)
print("Vectors propis", Vectors)
```

```python
u=Vectors[:,0] # posem aixo per extreure els vector propi de la primera columna
print(u)
```

comprovem que efectivament son vectors propis

```python
# per multiplicar matrius es la mteixa comnda que amb vectors. S'han de poder multiplicar. np.dot
z=np.dot(A,u)
print(z)
```

```python
v=Vectors[:,1] # posem aixo per extreure els vector propi de la segona columna
print(v)
```

```python
z2=np.dot(A,v)
print(z2)
```
